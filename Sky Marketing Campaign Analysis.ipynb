```python
# Further recommendations/analysis to support decision making - independence testing, conditional probabilities and odds ratios

# Chi-squared test for independence between upgrade state and campaigning for Movies Upgrade case (Treatment Group)
df_treat = data[data['control group'] == 'Treatment']
df_pivot = df_treat.pivot_table(index='Channel', columns='Movies Upgrade', values='control group', aggfunc=len)
print df_pivot
# We want to perform a Chi-squared test of Independence to see if upgrade and campaigns are independent
# First we calculate the degrees of freedom (2 is no. of upgrade levels/ 3 is no. of campaign levels)
print 'There are {} degrees of freedom\n'.format((2-1)*(3-1))

# Expected frequency Counts
e_11 = ((146419+124413+13681)*(146419+55))/(df_pivot.sum().sum())
e_12 = ((146419+124413+13681)*(124413+34))/(df_pivot.sum().sum())
e_13 = ((146419+124413+13681)*(13681+124))/(df_pivot.sum().sum())
e_21 = ((55+34+124)*(146419+55))/(df_pivot.sum().sum())
e_22 = ((55+34+124)*(124413+34))/(df_pivot.sum().sum())
e_23 = ((55+34+124)*(13681+124))/(df_pivot.sum().sum())
print 'Expected Frequency 11: {}'.format(e_11)
print 'Expected Frequency 12: {}'.format(e_12)
print 'Expected Frequency 13: {}'.format(e_13)
print 'Expected Frequency 21: {}'.format(e_21)
print 'Expected Frequency 22: {}'.format(e_22)
print 'Expected Frequency 23: {}\n'.format(e_23)

# Calculate the Chi-squared test statistic
chi_squared = (146419-e_11)/e_11 + (124413-e_12)/e_12 + (13681-e_13)/e_13 + (55-e_21)/e_21 + (34-e_22)/e_22 + (124-e_23)/e_23
print chi_squared
# Get P-value
p_value = 1 - stats.chi2.cdf(chi_squared, 2)
print 'P-value: {}'.format(p_value)
print 'Since the P-value ({}) is less than the significance level (0.05), we cannot accept the null hypothesis that upgrade state and campaign are independent'.format(round(p_value,4))
print ' Therefore, we conclude that there is a relationship between Upgrade state and Campaign type'
```

```python
# We have decided that Upgrade state an campaigns are related. We can think of 
# Campaign as a predictor and Upgrade state as a response
# The conditional probabilities are:
x1 = 55/(146419 + 55)
x2 = 34/(124413+34)
x3 = 124/(13681+124)
print 'The probability that a client upgrades under the email campaign is {}%'.format(round(x1,5)*100)
print 'The probability that a client upgrades under the online campaign is {}%'.format(round(x2,5)*100)
print 'The probability that a client upgrades under the outbound dialling campaign is {}%'.format(round(x3,5)*100)
```

```python
x = ['Email', 'Online', 'Outbound Dialling']
y = [x1, x2, x3]
barplot = sns.barplot(x, y)
barplot.set(xlabel = "Channel", ylabel = "Probability", title = "Conditional Probability that Client Upgrades Movies by Channel")
plt.show()
```

```python
# We can also calculate sample odds ratios by looking at any 2x2 table within the larger table
# Email vs. Online
theta_12 = round((124413 * 55) / (146419 * 34),4)
theta_12_odds = (round(theta_12,4) - 1) * 100
print 'The estimated value of {} means that clients contacted by the email campaign are {}% more likely to upgrade than clients contacted by the online campaign'.format(theta_12, theta_12_odds)
```
