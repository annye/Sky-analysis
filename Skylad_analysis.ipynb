```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
def p2f(x):
    return float(x.strip('%'))/100
```

```python
df['%'] = df['%'].apply(lambda x: p2f(x))
```

```python
# Filter out values less than zero
df = df[df['%'] >= 0]
```

```python
# Get campaign dataframes for % of Spend to Target
campaign_list = []
for campaign in set(df['Campaign Type']):
    x = df[df['Campaign Type'] == campaign]['%']
    tuple_campaign = (x, campaign)
    campaign_list.append(tuple_campaign)
```

```python
# Plot multiple histograms
n_bins = 10
x1 = campaign_list[0][0]
x2 = campaign_list[1][0]
x3 = campaign_list[2][0]
x4 = campaign_list[3][0]
x5 = campaign_list[4][0]

fig, axes = plt.subplots(nrows=3, ncols=2)
ax0, ax1, ax2, ax3, ax4, ax5 = axes.flatten()

ax0.hist(x1, n_bins, normed=1, histtype='bar')
ax0.set_title(campaign_list[0][1])

ax1.hist(x2, n_bins, normed=1, histtype='bar')
ax1.set_title(campaign_list[1][1])


ax2.hist(x3, n_bins, normed=1, histtype='bar')
ax2.set_title(campaign_list[2][1])


ax3.hist(x4, n_bins, normed=1, histtype='bar')
ax3.set_title(campaign_list[3][1])


ax4.hist(x5, n_bins, normed=1, histtype='bar')
ax4.set_title(campaign_list[4][1])

fig.tight_layout()
plt.show()
```
